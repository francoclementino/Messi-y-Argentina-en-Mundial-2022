# Pases de MESSI en la final del Mundo Qatar 2022.

```python
fig = plt.figure(figsize=(16, 9))
ax = fig.add_subplot(1, 1, 1)

pitch = VerticalPitch(
    pitch_type='statsbomb',
              goal_type='box',
              half=True
              )

pitch.draw(ax=ax)

pases_comp = pases_messi[~pases_messi['pass_outcome'].isin(['Incomplete', 'Pass Offside'])]
pases_imcomp = pases_messi[pases_messi['pass_outcome'].isin(['Incomplete', 'Pass Offside'])]

pitch.arrows(pases_comp['x'], pases_comp['y'], pases_comp['endX'], pases_comp['endY'], ax=ax, label='Pases Completos', color='green', width=2.5)
pitch.arrows(pases_imcomp['x'], pases_imcomp['y'], pases_imcomp['endX'], pases_imcomp['endY'], ax=ax, label='Pases Incompletos', color='red', width=2.5)

plt.legend()

ax.set_title("Pases de Messi en la final del mundo Qatar 2022")
```

# Apartado tiros de Argentina en todo el Mundial Qatar 2022

```python
match_ids = list(mundial[(mundial['home_team'] == 'Argentina') | (mundial['away_team'] == 'Argentina')].match_id.values)
match_ids
```

```python
tiros = pd.DataFrame()
```

```python
import pandas as pd

tiros = []

for i in match_ids:
    df = sb.events(match_id=i)
    df = df[df['type'] == 'Shot']
    tiros.append(df)

tiros = pd.concat(tiros, ignore_index=True)

```

```python
tiros[['x', 'y']] = tiros.location.apply(pd.Series)
tiros[['x', 'y', 'location']]
```

```python
tiros.possession_team.value_counts()
```

```python
tiros['shot_team'] = np.where(tiros['possession_team'] == 'Argentina', 'Argentina', 'No Argentina')
tiros.shot_team.value_counts()
```

```python
tiros_arg = tiros[tiros['shot_team'] == 'Argentina']
tiros_arg.shape
```

# Tiros de Argentina en los 7 partidos del Mundial 2022

```python
fig = plt.figure(figsize=(16, 9))
ax = fig.add_subplot(1, 1, 1)

pitch = VerticalPitch(pitch_type='statsbomb',
              goal_type='box',
              half=True
              )

pitch.draw(ax=ax)

plt.ylim(70, 125)

pitch.scatter(tiros_arg.x, tiros_arg.y, ax=ax, s=tiros_arg.shot_statsbomb_xg*500, alpha=.6, ec='black', label='Tiros')
plt.legend()
```

```python
goles_arg = tiros_arg[tiros_arg.shot_outcome == 'Goal']
no_goles_arg = tiros_arg[tiros_arg.shot_outcome != 'Goal']
goles_arg.shape, no_goles_arg.shape, tiros_arg.shape
```

# Ataque Argentina en el Mundial 2022

```python
fig = plt.figure(figsize=(16, 9))
ax = fig.add_subplot(1, 1, 1)

pitch = VerticalPitch(pitch_type='statsbomb',
              goal_type='box',
              half=True
              )

pitch.draw(ax=ax)

plt.ylim(70, 125)
plt.xlim(-5, 100)

pitch.scatter(goles_arg.x, goles_arg.y, ax=ax, s=goles_arg.shot_statsbomb_xg*500, alpha=.9, ec='black', label='Goles', color='blue', zorder=.99)
plt.legend()
pitch.scatter(no_goles_arg.x, no_goles_arg.y, ax=ax, s=no_goles_arg.shot_statsbomb_xg*500, alpha=.6, ec='black', label='Tiros', color='white', zorder=0)

plt.legend(ncol=2, loc='lower center', prop={'size' : 16}, shadow=True, edgecolor='black')

disparos = tiros_arg.shape[0]
goles = goles_arg.shape[0]
xG_total = tiros_arg.shot_statsbomb_xg.sum()
ax.text(96, 115, 'Ataque Argentina\n en el Mundial 2022', va='center', ha='center', fontsize=25)
ax.text(95, 105, f'{disparos} Tiros', va='center', ha='center', fontsize=20)
ax.text(95, 100, f'{goles} Goles', va='center', ha='center', fontsize=20)
ax.text(95, 95, f'{round(xG_total,2)} xG', va='center', ha='center', fontsize=20)
```

```python
tiros_arg.shot_statsbomb_xg.sum()
```

# Defensa Argentina en Mundial 2022

```python
fig = plt.figure(figsize=(16, 9))
ax = fig.add_subplot(1, 1, 1)

pitch = VerticalPitch(pitch_type='statsbomb',
              goal_type='box',
              half=True
              )

pitch.draw(ax=ax)

plt.ylim(70, 125)
plt.xlim(-5, 100)

tiros_no_arg = tiros[tiros['shot_team'] == 'No Argentina']

goles_no_arg = tiros_no_arg[tiros_no_arg.shot_outcome == 'Goal']
no_goles_no_arg = tiros_no_arg[tiros_no_arg.shot_outcome != 'Goal']

pitch.scatter(goles_no_arg.x, goles_no_arg.y, ax=ax, s=goles_no_arg.shot_statsbomb_xg*500, alpha=.9, ec='black', label='Goles', color='blue', zorder=.99)
plt.legend()
pitch.scatter(no_goles_no_arg.x, no_goles_no_arg.y, ax=ax, s=no_goles_no_arg.shot_statsbomb_xg*500, alpha=.6, ec='black', label='Tiros', color='white', zorder=0)

plt.legend(ncol=2, loc='lower center', prop={'size' : 16}, shadow=True, edgecolor='black')

disparos = tiros_no_arg.shape[0]
goles = goles_no_arg.shape[0]
xG_total = tiros_no_arg.shot_statsbomb_xg.sum()
ax.text(96, 115, 'Defensa Argentina\n en el Mundial 2022', va='center', ha='center', fontsize=25)
ax.text(95, 105, f'{disparos} Tiros', va='center', ha='center', fontsize=20)
ax.text(95, 100, f'{goles} Goles', va='center', ha='center', fontsize=20)
ax.text(95, 95, f'{round(xG_total,2)} xG', va='center', ha='center', fontsize=20)
```
